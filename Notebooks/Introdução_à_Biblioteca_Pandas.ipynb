```python
# Importando a Biblioteca Pandas
import pandas as pd
```

```python
# Criando uma Variável Data Frame
df = pd.read_csv("/content/Digital Innovation One/Gapminder.csv", error_bad_lines = False, sep = ";")
```

```python
# Visualizando as 5 Primeiras Linhas
df.head()
```

```python
# Renomeando as Colunas do Data Frame
df = df.rename(columns={"country" : "País", "continent" : "Continente", "year" : "Ano", "lifeExp" : "Expectativa de Vida", "pop" : "População Total", "gdpPercap" : "PIB"})
```

```python
# Visualizando as 10 Primeiras Linhas
df.head(10)
```

```python
# Total de Linhas e Colunas
df.shape
```

```python
# Visualizando as Colunas
df.columns
```

```python
# Visualizando o Tipo de Dados
df.dtypes
```

```python
# Visualizando as Últimas Linhas
df.tail(10)
```

```python
# Visualizando Informações Estatísticas do DataSet
df.describe()
```

```python
# Fazendo Filtro do Continente
df["Continente"].unique()
```

```python
# Fazendo Filtro para o Continente da Oceania
Oceania = df.loc[df["Continente"] == "Oceania"]
```

```python
# Visualizando a Informação do Filtro da Oceania as 5 primeiras Linhas
Oceania.head()
```

```python
# Fazendo Filtro do Continente Oceania
Oceania["Continente"].unique()
```

```python
# Agrupando Dados por Continente e País
df.groupby("Continente")["País"].nunique()
```

```python
# Agrupando Dados por Ano e Expectativa de Vida e Calculando a Média
df.groupby("Ano")["Expectativa de Vida"].mean()
```

```python
# Calculando a Média do PIB
df["PIB"].mean()
```

```python
# Calculando a Soma
df["PIB"].sum()
```
