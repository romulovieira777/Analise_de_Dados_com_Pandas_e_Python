# Tuplas

```python
# As tuplas usam parênteses como sintaxe
tp = ("Banana", "Maçã" ,10 ,50)
```

```python
# Retornando o primeiro elemento
tp[0]
```

```python
# Diferentes das listas as tuplas são imutáveis, o que quer dizer que não podemos alterar os seus elementos
tp[0] = "Laranja"
```

```python
# Contando quantas vezes aparece a palavra Maçã
tp.count("Maçã")
```

```python
# Retornando os dois primeiros elementos da tupla
tp[0:2]
```

# Dicionários

```python
# Para criar um dicionário utilizamos as {}
# OBS: Dicionários trabalham com chave e valor
dc = {"Maçã" : 20, "Banana" : 10, "Laranja" : 15, "Uva" : 5}
```

```python
# imprimindo o dicionário
dc
```

```python
# Acessando o valor de um dicionário através da chave
dc["Maçã"]
```

```python
# Atualizando o valor da Maçã
dc["Maçã"] = 25
```

```python
# Imprimindo o dicionário
dc
```

```python
# Retornando todas as chaves do dicionário
dc.keys()
```

```python
# Retornando os valores do dicionário
dc.values()
```

```python
# Verificando se já existe uma chave no dicionário e caso não exista inserir
dc.setdefault("Limão", 22)
```

```python
# Imprimindo o dicionário
dc
```
