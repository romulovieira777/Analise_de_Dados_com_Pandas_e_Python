# **Vizualização de Dados - Rômulo Vieira**


```python
# Importando a Biblioteca Pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Leitura dos Arquivos
df_01 = pd.read_excel("/content/Aracaju.xlsx")
df_02 = pd.read_excel("/content/Fortaleza.xlsx")
df_03 = pd.read_excel("/content/Natal.xlsx")
df_04 = pd.read_excel("/content/Recife.xlsx")
df_05 = pd.read_excel("/content/Salvador.xlsx")
```

```python
# Juntando todos os Arquivos
df = pd.concat([df_01, df_02, df_03, df_04, df_05])
```

```python
# Criando a Coluna de Receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
# Criando uma Nova Coluna com Ano
df["Ano_Venda"] = df["Data"].dt.year
```

```python
# Visualizando as informações do maior para o menor
df["LojaID"].value_counts(ascending=False)
```

```python
# Criando um gráfico de barras
df["LojaID"].value_counts(ascending=False).plot.bar()
```

```python
# Gráfico de barras horizontais
df["LojaID"].value_counts().plot.barh()
```

```python
# Gráfico de barras horizontais
df["LojaID"].value_counts(ascending=True).plot.barh();
```

```python
# Gráfico de pizza
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie()
```

```python
# Total vendas por cidade
df["Cidade"].value_counts()
```

```python
# Adicionando um título e alterando o nome dos eixos
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
# Alterando a cor
df["Cidade"].value_counts().plot.bar(title="Total vendas por Cidade", color="red")
plt.xlabel("Cidade")
plt.ylabel("Total Vendas");
```

```python
# Alterando o estilo
plt.style.use("ggplot")
```

```python
# Criando um gráfico ggplot
df.groupby(df["mes_venda"])["Qtde"].sum().plot(title = "Total Produtos Vendidos x Mês")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos");
plt.legend();
```

```python
# Agrupando por quantidade
df.groupby(df["mes_venda"])["Qtde"].sum()
```

```python
# Selecionando apenas as vendas de 2019
df_2019 = df[df["Ano_Venda"] == 2019]
```

```python
# Total produtos vendidos por mês
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum()
```

```python
# Total produtos vendidos por mês
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend();
```

```python
# Histograma
plt.hist(df["Qtde"], color = "magenta");
```

```python
# Criando um gráfico de dispersão
plt.scatter(x = df_2019["dia_venda"], y = df_2019["Receita"]);
```

```python
# Salvando em png
df_2019.groupby(df_2019["mes_venda"])["Qtde"].sum().plot(marker = "v")
plt.title("Quantidade de Produtos Vendidos x Mês")
plt.xlabel("Mês")
plt.ylabel("Total Produtos Vendidos")
plt.legend()
plt.savefig("GráficoQTDE x Mês.png");
```
